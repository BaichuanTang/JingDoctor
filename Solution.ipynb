```python
feat=features.nunique()
display(features.shape)

#过滤feature：
#把类型为Categorical且出现过多的features（这样OneHot后维度会爆炸，且多为id）和出现过少的features（没有区分性）过滤
features.drop(list(set(features.dtypes[features.dtypes=='object'].keys())&set(feat[(feat.values>1000)].keys())|set(feat[(feat.values<=1) ].keys())) ,axis=1,inplace=True)
display(features.shape)
```

```python
#分别对Categorical和Numerical做one hot，顺便得到变量名可解释，还有填充空值一步到位
display(Counter(features.dtypes.values))

obj_feat=features[features.dtypes[features.dtypes=='object'].keys()]
num_feat=features.drop(features.dtypes[features.dtypes=='object'].keys(),axis=1)
# enc = OneHotEncoder(categories='auto').fit(obj_feat) 
# obj_feat=enc.transform(obj_feat)
display(obj_feat.shape)
obj_feat=pd.get_dummies(obj_feat,dummy_na=True,drop_first=True)
for i in num_feat.columns:
    num_feat[i]=num_feat[i].fillna(num_feat[i].mean())
display(num_feat.shape)

num_cols=num_feat.columns
num_idxs=num_feat.index
#标准化
from sklearn import preprocessing
scaler = preprocessing.StandardScaler()
num_feat=scaler.fit_transform(num_feat)
num_feat=pd.DataFrame(num_feat,columns=num_cols,index=num_idxs)

all_feat=pd.concat([obj_feat,num_feat],axis=1)
display(all_feat.shape)

```

```python
#num_feat（其他也行）与y join，join后不为空为训练的部分，为空则是测试的数据
train_if_mat=num_feat.iloc[:,[0]].merge(y,left_index=True,right_on='客户ID',how='left')[['客户ID', '激活医美']].reset_index(drop=True)
train_matrix=all_feat.iloc[train_if_mat[~train_if_mat.激活医美.isna()].index]

test_matrix=all_feat.iloc[train_if_mat[train_if_mat.激活医美.isna()].index]
```

### 此处有风险：正确的方式应该是使用StratifiedKFold进行5折抽样，再对每一折的训练集Smote，验证集不Smote

我只调了max_depth和min_child_weight，这一步非常耗时！不要跑  
  
Xgboost参数调优的一般步骤：  
　　1、学习速率（learning rate）。在0.05~0.3之间波动，通常首先设置为0.1。  
　　2、进行决策树特定参数调优（max_depth , min_child_weight , gamma , subsample,colsample_bytree）在确定一棵树的过程中，我们可以选择不同的参数。  
　　3、正则化参数的调优。（lambda , alpha）。这些参数可以降低模型的复杂度，从而提高模型的表现。  
　　4、降低学习速率，确定理想参数。  
1、max_depth和min_child_weight  
首先设置固定的学习速率 learning rate，对max_depth、min_child_weight两个参数进行调优。  

```python
#max_depth和min_child_weight参数调优
# max_depth和min_child_weight参数对最终结果有很大的影响。max_depth通常在3-10之间，min_child_weight。采用栅格搜索（grid search），我们先大范围地粗略参数，然后再小范围的微调。
# 网格搜索scoring = 'roc_auc' 只支持二分类，多分类需要修改scoring（默认支持多分类）

param_test1 = {
'max_depth':[i for i in range(3,10,2)],
'min_child_weight':[i for i in range(1,6,2)]
}
from sklearn.model_selection import GridSearchCV
gsearch = GridSearchCV(
    estimator = XGBClassifier(
                learning_rate =0.01,
                n_estimators=500, 
                max_depth=5,
                min_child_weight=1,
                gamma=0,
                subsample=0.8,
                colsample_bytree=0.8,
                objective= 'binary:logistic',
                scale_pos_weight=1,
                seed=27),
                param_grid = param_test1,
                scoring='roc_auc',
                n_jobs=-1,
                cv=5)
gsearch.fit(X_train_os,y_train_os)
```

```python
print('gsearch1.grid_scores_', gsearch.best_estimator_)
print('gsearch1.best_params_', gsearch.best_params_)
print('gsearch1.best_score_', gsearch.best_score_)
print('gsearch1.best_score_', gsearch.cv_results_)
```

```python
import random
#单次划分训练验证，以找到最好轮数
random.seed(0)
L1 = random.sample(range(0, 9660), 7500)
L2=list(set(range(0, 9660))-set(L1))

from imblearn.over_sampling import SMOTE, ADASYN
smt=SMOTE(random_state=2)
X_train_os,y_train_os = smt.fit_resample(train_matrix.iloc[L1].values, list(y['激活医美'].iloc[L1]))

Counter(y_train_os).items()
```

```python
#用'max_depth': 3, 'min_child_weight': 5搭配其他参数，观察到模型在第6轮达到验证集最高AUC。（每次结果均不同，当时做的时候第六轮最好,也可以选更大的）
xgb_train = xgb.DMatrix(X_train_os,y_train_os)  
xgb_eval = xgb.DMatrix(np.array(train_matrix.iloc[L2]), list(y['激活医美'][L2]))  
params={
    'objective': 'binary:logistic',
    'eval_metric': 'auc',
    'max_depth': 3, 'min_child_weight': 5,
    'gamma': 0,
    'subsample': 0.8,
    'colsample_bytree': 0.8,
    'scale_pos_weight': 1,
    'learning_rate' : 0.01,
    'n_estimators': 800, 
    'seed':3,
    'silent':1}

watchlist = [(xgb_train,'train'),(xgb_eval,'test')]
bst=xgb.train(params,xgb_train,num_boost_round=100,early_stopping_rounds=20,evals=watchlist)
```
